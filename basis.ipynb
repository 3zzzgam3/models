**Variability Metrics**

*Deviations*

The difference between the observed values and the estimate of location.
Synonyms;errors, residuals

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
datos = np.array([10, 12, 23, 23, 16, 23, 21, 16])

# Calcular la media y la desviación estándar
media = np.mean(datos)

# Calcular las desviaciones respecto a la media
desviaciones = datos - media

# Mostrar resultados
print("Datos: ", datos)
print("Media: ", media)
print("Desviaciones respecto a la media: ", desviaciones)

# Graficar los datos y las desviaciones
plt.figure(figsize=(10, 5))
plt.plot(datos, 'bo-', label='Datos')
plt.axhline(y=media, color='r', linestyle='-', label='Media')
for i in range(len(datos)):
    plt.plot([i, i], [media, datos[i]], 'g--')

plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('Datos y Desviaciones respecto a la Media')
plt.legend()
plt.grid(True)
plt.show()
```

La varianza muestral es una estimación de la varianza poblacional basada en una muestra extraída de la población. Se denota comúnmente como 
𝑠^2 y se calcula con una corrección por los grados de libertad para evitar el sesgo en la estimación de la varianza poblacional. La fórmula es:



La razón para usar 𝑛−1 en lugar de 𝑛 en la fórmula de la varianza muestral es que estamos estimando la media poblacional con la media muestral, y esto introduce un sesgo que se corrige dividiendo por 𝑛−1 en lugar de 𝑛


```python
import numpy as np

# Datos de ejemplo (una muestra de una población)
datos = np.array([10, 12, 23, 23, 16, 23, 21, 16])

# Calcular la media de la muestra
media_muestral = np.mean(datos)

# Calcular la varianza muestral (con n-1 grados de libertad)
varianza_muestral = np.var(datos, ddof=1)

# Calcular la varianza poblacional (asumiendo que los datos representan la población completa)
varianza_poblacional = np.var(datos, ddof=0)

# Mostrar resultados
print("Datos: ", datos)
print("Media muestral: ", media_muestral)
print("Varianza muestral: ", varianza_muestral)
print("Varianza poblacional: ", varianza_poblacional)
```

La desviación estándar y la desviación estándar muestral son medidas de dispersión que indican cuánto varían los datos respecto a la media. Aunque conceptualmente similares, se calculan de manera diferente dependiendo de si estamos considerando una población completa o una muestra.

La desviación estándar muestral se utiliza cuando se tiene una muestra de la población y se quiere estimar la desviación estándar de la población. Se denota como 𝑠 y se calcula con una corrección de grados de libertad para evitar sesgo:



```python
import numpy as np

# Datos de ejemplo (una muestra de una población)
datos = np.array([10, 12, 23, 23, 16, 23, 21, 16])

# Calcular la media de la muestra
media_muestral = np.mean(datos)

# Calcular la desviación estándar muestral (con n-1 grados de libertad)
desviacion_estandar_muestral = np.std(datos, ddof=1)

# Calcular la desviación estándar poblacional (asumiendo que los datos representan la población completa)
desviacion_estandar_poblacional = np.std(datos, ddof=0)

# Mostrar resultados
print("Datos: ", datos)
print("Media muestral: ", media_muestral)
print("Desviación estándar muestral: ", desviacion_estandar_muestral)
print("Desviación estándar poblacional: ", desviacion_estandar_poblacional)
```

**Variance and standar deviation**



```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo (una muestra de una población)
datos = np.array([10, 12, 23, 23, 16, 23, 21, 16])

# Calcular la media de la muestra
media_muestral = np.mean(datos)

# Calcular la varianza y desviación estándar muestral (con n-1 grados de libertad)
varianza_muestral = np.var(datos, ddof=1)
desviacion_estandar_muestral = np.std(datos, ddof=1)

# Calcular la varianza y desviación estándar poblacional (asumiendo que los datos representan la población completa)
varianza_poblacional = np.var(datos, ddof=0)
desviacion_estandar_poblacional = np.std(datos, ddof=0)

# Mostrar resultados
print("Datos: ", datos)
print("Media muestral: ", media_muestral)
print("Varianza muestral: ", varianza_muestral)
print("Desviación estándar muestral: ", desviacion_estandar_muestral)
print("Varianza poblacional: ", varianza_poblacional)
print("Desviación estándar poblacional: ", desviacion_estandar_poblacional)

# Graficar los datos y las desviaciones estándar
plt.figure(figsize=(12, 6))

# Graficar los datos
plt.plot(datos, 'bo-', label='Datos')

# Línea de la media
plt.axhline(y=media_muestral, color='r', linestyle='-', label='Media')

# Rango de la desviación estándar muestral
plt.axhline(y=media_muestral + desviacion_estandar_muestral, color='g', linestyle='--', label='Media + 1 Desviación Estándar Muestral')
plt.axhline(y=media_muestral - desviacion_estandar_muestral, color='g', linestyle='--', label='Media - 1 Desviación Estándar Muestral')

# Rango de la desviación estándar poblacional
plt.axhline(y=media_muestral + desviacion_estandar_poblacional, color='y', linestyle='-.', label='Media + 1 Desviación Estándar Poblacional')
plt.axhline(y=media_muestral - desviacion_estandar_poblacional, color='y', linestyle='-.', label='Media - 1 Desviación Estándar Poblacional')

plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('Datos, Media y Desviaciones Estándar')
plt.legend()
plt.grid(True)
plt.show()
```
